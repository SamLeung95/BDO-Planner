```python
import xlrd
import csv
```

```python
def main(index):
    f=open('armors.csv','rt')
    reader=csv.reader(f)
    items =[]
    for row in reader:
        items.append(row)
   
    printItem(index, items)
        
def printItem(index, items):
    gear=[]
    
    for item in items:
        if(item[0]==index):
            gear.append(item)

    print ('			\"ev\": ' + gear[0][51] + ',')
    print ('			\"hev\": ' + gear[0][52] + ',')
    print ('			\"dr\": ' + gear[0][53] + ',')
    print ('			\"hdr\": ' + gear[0][54])
    print ('		},')
    print ('		\"enhancement_text\": \"\",')
    print ('		\"enhancement\": {')
    for item in gear:
        if(item[2]!='0' and item[2]!='20'):
            dp = int(item[51])+ int(item[53])
            print ('			\"' + item[2] + '\": {')
            print ('				\"dp\": ' + str(dp) + ',')
            print ('				\"ev\": ' + item[51] + ',')
            print ('				\"hev\": ' + item[52] + ',')
            print ('				\"dr\": ' + item[53] + ',')
            print ('				\"hdr\": ' + item[54])
            print ('			},')
            
    dp = int(gear[20][51])+ int(gear[20][53])
    print ('			\"' + gear[20][2] + '\": {')
    print ('				\"dp\": ' + str(dp) + ',')
    print ('				\"ev\": ' + gear[20][51] + ',')
    print ('				\"hev\": ' + gear[20][52] + ',')
    print ('				\"dr\": ' + gear[20][53] + ',')
    print ('				\"hdr\": ' + gear[20][54])
    print ('			}')
    print ('		},')
```
